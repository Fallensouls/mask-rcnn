```python
import pandas as pd

df = pd.read_excel('../data/food_segm/fooddata_sort.xlsx')
data=df.values
# print("获取到所有的值:\n{}".format(data))
data = [str(x) for x in data][:-10]
print(data)
rows = []
for i in range(len(data)):
    d = data[i].split(" ")
    row = [d[0], d[1], d[3], d[5]]
    row = [int(i[1:]) for i in row]
    rows.append(row)
print(rows)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

path = '../data/food_segm/colors/1.jpeg'
# path = '../data/food_images/1.png'
image = cv2.imread(path, cv2.IMREAD_UNCHANGED)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# print(image)
# image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
np.set_printoptions(threshold=np.inf)
# print(gray)
plt.imshow(image)
plt.show()
# obj_ids = np.unique(gray)
# print(obj_ids)
# print(gray)
index = np.where(gray<70)
index2 = np.where((gray > 80) & (gray < 110))
index3 = np.where((gray > 140) & (gray < 190))
gray[index] = 1
gray[index2] = 2
gray[index3] = 3
# _, binary_mask = cv2.threshold(gray.astype(np.uint8), 70, 255, cv2.THRESH_BINARY)
plt.imshow(gray)
plt.savefig('mask.jpg', dpi=500)
plt.show()
```
